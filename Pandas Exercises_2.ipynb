```python
import pandas as pd

##import data csv format
all_data=pd.read_csv("https://raw.githubusercontent.com/KeithGalli/Pandas-Data-Science-Tasks/master/SalesAnalysis/Output/all_data.csv")

## data cleaning
## drop null columns
all_data = all_data.dropna(how='all')

all_data = all_data[all_data['Quantity Ordered'].str[0:2]!='Qu']
all_data = all_data[all_data['Order Date'].str[0:2]!='Or']

all_data['Month'] = all_data['Order Date'].str[0:2]

all_data['Month'] = all_data['Month'].astype('int32')

all_data['Quantity Ordered']=all_data['Quantity Ordered'].astype('int32')
## Quantity order mean by Product
all_data.groupby('Product')['Quantity Ordered'].mean()

## add column

all_data['NewColumn']=pd.to_datetime( all_data['Order Date']).dt.year
all_data

## Write a function and get only city name from adress column
def GetCityName(inputAdress):
    return inputAdress.split(",")[1]

all_data['CityName']=all_data['Purchase Address'].apply(GetCityName)

## add new column as Sales and show Quantity * product


all_data['Sales']=all_data['Quantity Ordered'].astype(int) * all_data['Price Each'].astype('float')


##grouping by Sales

all_data.groupby(by='Month')['Sales'].sum()


##visiulation

import matplotlib.pyplot as plt

plt.bar(range(1,13),all_data.groupby(by='Month')['Sales'].sum())
plt.xlabel="Months"
plt.ylabel="Sales by Month"
plt.show()
```
