```python
import requests as req 
import pandas as pd
import json
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
url = "https://github.com/ethen8181/machine-learning/blob/master/python/cohort/relay-foods.csv"
df = pd.read_csv(url + "?raw=true")
df=df[['OrderDate','UserId','TotalCharges']]
df['OrderDate']=pd.to_datetime(df['OrderDate'])
df['TotalCharges']=(df['TotalCharges'].apply(lambda x:x.replace('$',''))).astype(float)
get_year_month=lambda x:datetime(x.year,x.month,1)
df['OrderMonth']=df['OrderDate'].apply(get_year_month)

df['CohortMonth']=df.groupby('UserId')['OrderDate'].transform(min)
df['CohortMonth']=df['CohortMonth'].apply(lambda x:datetime(x.year,x.month,1))
years_diff=df['OrderMonth'].dt.year-df['CohortMonth'].dt.year
month_diff=df['OrderMonth'].dt.month-df['CohortMonth'].dt.month
df['cohortIndex']=years_diff*12+month_diff*1
grouping_count=df.groupby(['CohortMonth','cohortIndex'])
cohort_data = df.groupby(['CohortMonth','cohortIndex'])['UserId'].apply(pd.Series.nunique)
cohort_data = cohort_data.reset_index()
cohort_user_counts = cohort_data.pivot(index='CohortMonth',
                                  columns='cohortIndex',
                                  values='UserId')

cohort_sizes = cohort_user_counts.iloc[:,0]
retention = cohort_user_counts.divide(cohort_sizes, axis=0)
retention
```
