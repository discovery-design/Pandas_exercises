```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime

url="https://raw.githubusercontent.com/KeithGalli/Data-Science-Project-Ideas/master/Covid/time_series_covid19_confirmed_global.csv"

data=pd.read_csv(url)

data_grouped=data.groupby(by='Country/Region').sum()

df = data_grouped.drop(['Lat', 'Long'], axis=1).reset_index()

##data_grouped_dropped[data_grouped_dropped['Country/Region']=='Turkey']

countries=['US', 'China', 'United Kingdom', 'Germany', 'India']


start_date='01/22/2020'
end_date='04/03/2020'


def GetDayDifferenceBetweenTwoDates(date_string):



    date_list=[int(x) for x in date_string.split('/')]

    dateObject=datetime.datetime(date_list[-1],date_list[0],date_list[1])

    dateObjectOfStart=datetime.datetime(2020,1,22)

    return (dateObject-dateObjectOfStart).days




dates = list(df.columns[1:])

start_index = GetDayDifferenceBetweenTwoDates(start_date)
end_index = GetDayDifferenceBetweenTwoDates(end_date)

x_values = dates[start_index:end_index+1]

for country in countries:
    country_df = df[df['Country/Region'] == country]
    y_values = [int(country_df[col]) for col in x_values]
    plt.plot(x_values, y_values, label=country)



plt.legend()
plt.show()


```
