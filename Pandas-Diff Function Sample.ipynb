```python

## Her pokemonun Defense değerinin bir önceki pokemona göre yüzde kaç fazla veya eksik olduğunu bulalım
##Bulduğumuz değerleri YuzdelikDegisim şeklinde yeni bir kolon açıp o kolona yazalım


import pandas as pd
pokemon_url='https://raw.githubusercontent.com/rpdieego/IndigoTortoise-pkmn-notebooks/master/datasets_2619_4359_pokemon.csv'
pokemon_df=pd.read_csv(pokemon_url)
pokemon_df['DefenseDiff']=pokemon_df['Defense'].diff()
pokemon_df['OncekiDeger']=pokemon_df['Defense']-pokemon_df['DefenseDiff']
pokemon_df['YuzdelikDegisim']=pokemon_df['DefenseDiff']/pokemon_df['OncekiDeger']
pokemon_df
```
